**Approximating $\sin{x}$**

In this activity, we will use the **Taylor Series** to approximate the function $\sin{x}$. 

The general definition of the Taylor series for expandinf $f(x)$ about the point $x = x_0$ is:
$$f(x) = \sum_{n=0}^{\infty} \frac{f^{(j)}(x_o)}{j!}(x-x_0)^j $$

**We will visualize the effect of adding more terms to this series.** We will do this by expanding $f(x) = \sin{x}$ about the point $x = \pi$.
