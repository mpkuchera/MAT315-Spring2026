**Forward difference:**
  $$  f'(x) \approx \frac{f(x+h) - f(x)}{h} $$

**Central difference:**
    $$ f'(x) \approx \frac{f(x+h) - f(x-h)}{2h} $$
    

From Taylor expansion we found:
 - Forward difference error:  ~ $\mathcal{O}(h^2)$
 - Central difference error:  ~$\mathcal{O}(h^3)$
